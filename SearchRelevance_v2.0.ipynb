```python
#Finding tfidf for column query
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_cnv=TfidfVectorizer(stop_words='english',min_df=0.0001,analyzer='word')
search_str=tfidf_cnv.fit_transform(search_df['query'])
```

```python
#Removing numbers from and punctuations from the title field
import re
def clean_data(text):
    text=re.sub(r'\d+(?:\.\d*(?:[eE]\d+))?','NUMBER', text)
    text=re.sub(r'\W+',' ',text,flags=re.M)
    return text
```

```python
search_title=search_df['product_title'].apply(clean_data)
```

```python
search_title
```

```python
#Tfidf on Title column
tfidf_cnv=TfidfVectorizer(stop_words='english',min_df=0.0001,analyzer='word')
search_title_mat=tfidf_cnv.fit_transform(search_title)
```

```python
search_title_mat
```

```python
search_str
```

```python
#Finding cosine distance between query and product title
from sklearn.metrics.pairwise import cosine_similarity
distance=cosine_similarity(search_str,search_title_mat)
```
